# Individual Correlations

```python
'''

function for generating the correlation to overall score for each individual test. Takes 2 inputs:
    x = regex string to identify the test
    d = overall dataframe containing the test
function returns a df of correlations to overall score after filtering out any score with low well being.


'''
def test_corr(x, d):
    # filter dataframe to regex string x
    df1 = d.filter(regex = x) #to handle any NA values
    
    #generate overall score
    df1['overall_score'] = df1.mean(axis = 1)
    
    # calculate cutoff point for wellbeing
    max_l = df1.max().to_list()
    cut_off = max(max_l)/2
    
    # filter dataframe to above wellbeing cutoff
    df1_filtered = df1[df1['overall_score'] > cut_off]
    
    #drop overall score
    df1_filtered.drop('overall_score', axis=1, inplace = True)    
    
    # run correlation 
    corr_df = df1_filtered.corr()
    
    # update correlation values of 1 and negative to reflect 0s
    corr_df = corr_df.mask(corr_df == 1, 0)
    corr_df = corr_df.mask(corr_df < 0, 0)
    

    # return dict
    return corr_df
```

```python
#I think we should remove qlesq due to the med/no med consideration

test_list = ['^mansa_\d', '^whoqol', '^eq5d', '^who5', '^k6', '^pgwbi', 'sf36']

name_list = ['mansa', 'whoqol', 'eq5d', 'who5', 'k6', 'pgwbi', 'sf36']

for i in range(len(test_list)):
    globals()[name_list[i]] = test_corr(test_list[i], df)
```

# Overall Correlation

```python
'''

function for generating the correlation to overall score for each individual test. Takes 2 inputs:
    x = regex string to identify the test
    d = overall dataframe containing the test
function returns a df of correlations to overall score after filtering out any score with low well being.


'''
def loop_corr(d):
    
    
    #create empty dataframe to which all columns will be appended after they are individually filtered (by QOL)
    #df1 = d.filter(regex='|'.join(test_list))
    df2 = pd.DataFrame()
    
    #regex list
    test_list = ['^mansa_\d', '^whoqol', '^eq5d', '^who5', '^k6', '^pgwbi', 'sf36']
    #test_list = ['^mansa_\d']
    
    #run for all quality of life tests
    for regex_phrase in test_list:
        # filter dataframe to regex string x
        df1 = d.filter(regex = regex_phrase)
    
        #generate overall score
        df1['overall_score'] = df1.mean(axis = 1)
    
        # calculate cutoff point for wellbeing
        max_l = df1.max().to_list()
        cut_off = max(max_l)/2
    
        # filter dataframe to above wellbeing cutoff
        #This will create different size dataframes for each quality of life score group
        #Wdf1_filtered = df1[df1['overall_score'] > cut_off]
        
        #Just replace ineligible scores with np.nan as it 
        df1_filtered = df1.where(df1['overall_score'] > cut_off, np.nan)
        
    
        #drop overall score
        df1_filtered.drop('overall_score', axis=1, inplace = True)
        
        #join the intermediate results to the overall dataframe
        #join works if the first dataframe isn't empty, append only works if the first one is: see samples below
        #df2 = df2.join(df1_filtered)
        #df2 = df2.append(df1_filtered)
        df1_columns = df1_filtered.columns
        df2[df1_columns] = df1_filtered
        
    
    # run correlation 
    corr_df = df2.corr()
    
    # update correlation values of 1 and negative to reflect 0s
    corr_df = corr_df.mask(corr_df == 1, 0)
    corr_df = corr_df.mask(corr_df < 0, 0)
    
    
    #Update correlation values of NaN to reflect 0s
    #Check samples down below
    corr_df = corr_df.mask(corr_df.isna(), 0)

    # return dataframe of correlation, and df2 for checking
    return corr_df, df2
```
